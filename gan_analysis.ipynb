```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load traing metrics data
metrics = pd.read_json("checkpoints/training_metrics.json")

#plot the generator and discriminator loss
plt.plot(metrics["G_loss"], label="G loss")
plt.plot(metrics["D_loss"], label="D loss")
plt.legend()
plt.xlabel("Epoch")
plt.ylabel("Loss")

plt.show()
```
